#### Count of orders 

#### By Region 

```python
# count of order by Region
s_data.groupby('Region')['Region'].count()
```

```python
#Sorting Data in decending order (Region)
region_sort = s_data.groupby(['Region'])['Order ID'].count().reset_index(name = 'Count').sort_values(['Count'], ascending = False)
region_sort
```

```python
region_sort.plot(kind='bar',x='Region',y='Count',color='red')
```

```python
region_sort.reset_index()

#Top 5 Regions where most orders are placed
t_5 = region_sort.head()
t_5
```

```python
t_5.plot(kind='bar',x='Region',y='Count',color='red')
```

```python
#Bottom 5 Regions which have most orders
b_5 = region_sort.tail()
b_5
```

```python
b_5.plot(kind='bar',x='Region',y='Count',color='red')
```

#### By Segment 

```python
# count of order by Segment
s_data.groupby('Segment')['Segment'].count()
```

```python
seg_sort = s_data.groupby(['Segment'])['Order ID'].count().reset_index(name = 'Count').sort_values(['Count'], ascending = False)
```

```python
seg_sort.plot(kind = 'line',x='Segment',y='Count',color='red')
```

#### By Category 

```python
# count of order by Category
s_data.groupby('Category')['Category'].count()
```

```python
cat_sort = s_data.groupby(['Category'])['Order ID'].count().reset_index(name = 'Count').sort_values(['Count'], ascending = False)
```

```python
cat_sort.plot(kind = 'line',x='Category',y='Count',color='red')
```
